```python
df_combined = pd.concat([df_Pepsi, df_Coke], axis=1)
print(df_combined)
```

```python
columns = df_combined.columns.to_list()


labels = ["Year Considered", "First Quarter Revenue (in Millions)", 
          "Last Quarter Revenue (in Millions)", "Total Revenue (in Millions)"]


pepsi_count = 0
coke_count = 0


new_columns = []

for col in columns:
    if col in labels:
        if pepsi_count < len(labels):  
            new_columns.append(f"Pepsi_{col.replace(' (in Millions)', '').replace(' ', '_')}")
            pepsi_count += 1
        else:
            new_columns.append(f"Coke_{col.replace(' (in Millions)', '').replace(' ', '_')}")
            coke_count += 1
    else:
        new_columns.append(col)  

# Assigning new column names
df_combined.columns = new_columns

# Display
print(df_combined.head())

```

```python

# Define the independent and dependent variables for Pepsi
X_pepsi = df_combined[['Pepsi_First_Quarter_Revenue']]  # Predictor (Q1 revenue)
Y_pepsi = df_combined['Pepsi_Last_Quarter_Revenue']  # Target (Q4 revenue)

# Add a constant for the intercept in the regression model
X_pepsi = sm.add_constant(X_pepsi)

# Create and fit the regression model for Pepsi
model_pepsi = sm.OLS(Y_pepsi, X_pepsi).fit()

# Print Pepsi regression summary
print("Pepsi Regression Results:")
print(model_pepsi.summary())

# Define the independent and dependent variables for Coke
X_coke = df_combined[['Coke_First_Quarter_Revenue']]  # Predictor (Q1 revenue)
Y_coke = df_combined['Coke_Last_Quarter_Revenue']  # Target (Q4 revenue)

# Add a constant for the intercept
X_coke = sm.add_constant(X_coke)

# Create and fit the regression model for Coke
model_coke = sm.OLS(Y_coke, X_coke).fit()

# Print Coke regression summary
print("\nCoca-Cola Regression Results:")
print(model_coke.summary())



########################################################################################################

## Outputs:

## Intercept (const): The baseline value when Q1 revenue is 0.

## Slope (coef): How much Q4 revenue changes per unit increase in Q1 revenue.

## R-squared: Shows how well Q1 revenue explains Q4 revenue (closer to 1 = stronger relationship).
     # Pepsi/Coke Q1 revenue is a strong predictor of Q4 revenue.

## P-values: Indicate statistical significance (p < 0.05 suggests a strong relationship).
```

```python
# Calculate year-over-year percentage change
df_combined['Pepsi_Revenue_Growth'] = df_combined['Pepsi_Total_Revenue'].pct_change() * 100
df_combined['Coke_Revenue_Growth'] = df_combined['Coke_Total_Revenue'].pct_change() * 100

# Plot the revenue growth over the years
plt.figure(figsize=(8, 5))
plt.plot(df_combined['Pepsi_Year_Considered'], df_combined['Pepsi_Revenue_Growth'], marker='o', label='Pepsi', color='blue')
plt.plot(df_combined['Coke_Year_Considered'], df_combined['Coke_Revenue_Growth'], marker='o', label='Coca-Cola', color='red')

plt.axhline(y=0, color='gray', linestyle='--')
plt.xlabel("Year")
plt.ylabel("Yearly Revenue Growth (%)")
plt.title("Yearly Revenue Growth of Pepsi vs. Coca-Cola")
plt.legend()
plt.show()
```

```python
# Rename the necessary columns for simplicity
df_combined = df_combined.rename(columns={
    'Pepsi_First_Quarter_Revenue': 'Pepsi_Q1',
    'Pepsi_Last_Quarter_Revenue': 'Pepsi_Q4',
    'Coke_First_Quarter_Revenue': 'Coke_Q1',
    'Coke_Last_Quarter_Revenue': 'Coke_Q4'
})

plt.figure(figsize=(8, 6))

# Scatter Plot
sns.scatterplot(x=df_combined['Coke_Q1'], y=df_combined['Coke_Q4'], color='red', label='Coca-Cola')
sns.scatterplot(x=df_combined['Pepsi_Q1'], y=df_combined['Pepsi_Q4'], color='blue', label='Pepsi')

# Fit regression line using statsmodels
for company, color in [('Coke', 'red'), ('Pepsi', 'blue')]:
    X = df_combined[f'{company}_Q1']
    Y = df_combined[f'{company}_Q4']
    X = sm.add_constant(X)  # Add constant for regression
    model = sm.OLS(Y, X).fit()
    predicted = model.predict(X)
    plt.plot(df_combined[f'{company}_Q1'], predicted, color=color, linestyle='--')

plt.xlabel("First Quarter Revenue")
plt.ylabel("Last Quarter Revenue")
plt.title("Q1 vs Q4 Revenue with Regression Line")
plt.legend()
plt.show()





########################################################################################################

# Interpretation 

## Positive Slope: If the regression line is upward-sloping, it means that when Q1 revenue increases, Q4 revenue also tends to increase.

## Flatter Line: If the regression line is flatter, it suggests a weaker relationship between Q1 and Q4 revenue.

## Both companies follow similar revenue trends but Pepsis line is steeper than Cokes, so Pepsis Q4 revenue is more strongly tied to Q1 revenue.
```
