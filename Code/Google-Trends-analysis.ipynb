```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
file_path = "/Users/jennanowland/Desktop/Data_400/Data400_Final/Data-Files/updated_googletrends.csv"
df = pd.read_csv(file_path)

df
```

```python
df['Month'] = pd.to_datetime(df['Month'])

# Filter Pepsi data for months 1, 2, 3
Pepsi_df = df[df['Month'].dt.month.isin([1, 2, 3])][['Month', 'Pepsi: (United States)']]

# Filter Coca-Cola data for months 11, 12
Coke_df = df[df['Month'].dt.month.isin([11, 12])][['Month', 'Coca-Cola: (United States)']]

# Display filtered DataFrames
print("Pepsi_df:")
print(Pepsi_df)
print("\nCoke_df:")
print(Coke_df)
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Convert 'Month' column to datetime format for better plotting
Pepsi_df['Month'] = pd.to_datetime(Pepsi_df['Month'])

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(Pepsi_df['Month'], Pepsi_df['Pepsi: (United States)'], marker='o', color='blue', label='Frequency')

# Labels and title
plt.xlabel("Month")
plt.ylabel("Frequency")
plt.title("Pepsi Google Searches Over Time (Jan-March)")

# Formatting x-axis
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Convert 'Month' column to datetime format
Coke_df['Month'] = pd.to_datetime(Coke_df['Month'])

# Convert 'Coca-Cola: (United States)' to numeric, replacing "<1" with 0.5
Coke_df['Coca-Cola: (United States)'] = Coke_df['Coca-Cola: (United States)'].replace('<1', 0.5).astype(float)

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(Coke_df['Month'], Coke_df['Coca-Cola: (United States)'], marker='o', color='red', label='Frequency')

# Labels and title
plt.xlabel("Month")
plt.ylabel("Frequency")
plt.title("Coke Google Searches Over Time (Nov-Dec)")

# Formatting x-axis
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(Coke_df['Month'], Coke_df['Coca-Cola: (United States)'], marker='o', color='red', label='Coca-Cola')
plt.plot(Pepsi_df['Month'], Pepsi_df['Pepsi: (United States)'], marker='o', color='blue', label='Pepsi')

plt.xlabel("Month")
plt.ylabel("Frequency")
plt.title("Coca-Cola vs. Pepsi Google Searches Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
